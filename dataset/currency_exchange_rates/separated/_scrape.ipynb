```python
# NOTE: to make sure the script is functioning correctly, 
# make sure to delete all csv files in current directory, 
# restart to Jupyter Notebook and clear all output, and 
# then run all cells
import yfinance as yf
import pandas as pd
import os
import glob
```

```python
# read in metadata
metadata = pd.read_csv("../_metadata.txt")
metadata
```

```python
# handling each csv's col adding, re-naming, & format adjustion
path = "./*.csv"
csv_files = glob.glob(os.path.join(path, "*.csv"))
for fname in glob.glob(path):
    df = pd.read_csv(fname)
    df.columns.values[0] = "Date" # rename date col
    df.columns.values[1] = "Exchange_rate_against_USD" # rename exchange rate col
    symbol = fname.split("\\")[1].split(".")[0]
    df["Symbol"] = symbol # add symbol col
    # add name col
    df["Name"] = metadata.loc[metadata["Symbol"] ==
                              symbol, "Name"].to_string(index=None).strip()
    # Re-adjust column sequence to: Name, Symbol, Date, exchange_rate
    df = df.reindex(columns=["Name", "Symbol", "Date", "Exchange_rate_against_USD"])
    # adjust the date format to: yyyy-mm-dd
    df["Date"] = pd.to_datetime(df["Date"])
    df.to_csv(fname, index=None)
```

```python
# merge into one csv

all_files = glob.glob(path)
df_merged = pd.concat(map(pd.read_csv, all_files), ignore_index=True)
df_merged.to_csv("../_exchangeRate_merged.csv", index=None)
```
